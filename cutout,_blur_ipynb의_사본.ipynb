```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt

image = cv2.cvtColor(cv2.imread('/content/겨울 쿨톤 연예인 화보2.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#Image Augmentation을 편하게 해주는 imgaug 라이브러리
#https://uos-deep-learning.tistory.com/17
```

```python
import imgaug.augmenters as iaa

aug = iaa.Cutout(nb_iterations=1, size = 0.4) # 개수랑 사이즈 조절할 수 있음!
image_aug = aug(image = image)

plt.imshow(image_aug)
```

```python
###########블러처리#########
#가우시안 블러링 https://blog.naver.com/PostView.naver?blogId=sees111&logNo=222366804864
```

```python
def blur(image):
    aug2 = iaa.GaussianBlur(sigma=0.9) #  GaussianBlur #sigma 값을 키울수록 blur정도가 높아
    return aug2(image=image) # GaussianBlur 적용 후 결과 반환

image_aug = blur(image = image) # 인자 이름을 image로 설정(단일 이미지 적용)

plt.imshow(image_aug)
```
